```python
# Question 1


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Setting random seed for reproducibility
np.random.seed(30)

# Generating data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Creating DataFrame
df = pd.DataFrame(data)

# 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
#  2 voilin plot


import seaborn as sns


# Seed and data generation
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df)
plt.title('Distribution of Scores Across Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()

```

```python
#  3  heat map


# Seed and data generation
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
# Question 5 



import plotly.express as px

# Seed for reproducibility
np.random.seed(25)

# Dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create bubble chart
fig = px.scatter(df, x='GDP', y='Population', size='Population', color='Country',
                 hover_name='Country', size_max=60, title='Country Population vs GDP')

# Show the plot
fig.show()
```
