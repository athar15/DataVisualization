```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Question 1


# Set the random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n = 100  # Number of data points
x = np.linspace(0, 10, n)
y = 2.5 * x + np.random.normal(0, 1, n)  # Linear relationship with some noise

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})
```

```python
data
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', color='r', marker='o', edgecolor='black')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title('Scatter Plot of Synthetic Data')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Question 2


# Generate random data
np.random.seed(0)  # Set seed for reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)

# Create a histogram using seaborn
sns.histplot(data, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Random Numbers')
plt.show()
```

```python
# Queston 3

# Step 1: Create a dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [23, 45, 56, 78, 12]
}
df = pd.DataFrame(data)

# Step 2: Visualize the data using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Bar plot
sns.barplot(x='Category', y='Value', data=df, palette='viridis')

# Add titles and labels
plt.title('Comparison of Categories Based on Numerical Values')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.show()
```

```python
# Question 4

# Step 1: Create a dataset
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
    'Value': [23, 45, 56, 78, 12, 34, 67, 89, 45, 23, 56, 78]
}
df = pd.DataFrame(data)

# Step 2: Visualize the data using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Box plot
sns.boxplot(x='Category', y='Value', data=df, palette='viridis')

# Add titles and labels
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.show()
```

```python
# Question 5


# Step 1: Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
size = 1000

# Generate random data
x1 = np.random.normal(0, 1, size)
x2 = 0.5 * x1 + np.random.normal(0, 1, size)  # Correlated with x1
x3 = -0.2 * x1 + np.random.normal(0, 1, size)  # Correlated with x1
x4 = np.random.normal(0, 1, size)  # Not correlated with x1

# Create a DataFrame
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Step 2: Visualize the correlation matrix using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)

# Add titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```
